# Función Senoidal

Probando multiples senoidales de distintas frecuencias y fases:
- $N = 1000$
- $f_s = 1000 [Hz]$

```python
import sys
import os

sys.path.append(os.path.abspath(".."))
from pds_utils import sine, square, plotter

import numpy as np

# (frequency, phase)
sines = [
    (50.0, np.pi),
    (100.0, np.pi / 4.0),
    (250.0, 3.0 / 4.0 * np.pi),
    (500.0, 0.0),
    (500.0, np.pi / 5),
    (500.0, np.pi / 2),
]
sine.plot_multiple_sine(sampling_freq=1000.0, samples=1000, amplitude=1, frequencies_and_phase=sines)
```

# Implementación de Funciones individuales 
- seno
- square-wave

Ambas señales utilizan los siguientes parámetros:
- $f = 100 [Hz]$
- $N = 1000$
- $f_s = 40 [kHz]$
- $\text{amplitude} = 1$
  
Por último, se gráfican ambas señales en color azul, y los puntos de "muestra" (sampling points) en rojo con un `scatter-plot`.

```python
f = 100
N = 1000
fs = 40e3
a = 1

t1, s1 = square.wave(sampling_freq=fs, samples=N, amplitude=a, dc_level=-1, frequency=f)
t2, s2 = sine.wave(sampling_freq=fs, samples=N, amplitude=a, dc_level=1, frequency=f)

plotter.multiple([(t1, s1, f), (t2, s2, f)])
```
